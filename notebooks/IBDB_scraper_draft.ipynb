```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as BS
from IPython.core.display import HTML
```

### scraping list of theatres

```python
theatre_URL = 'https://www.ibdb.com/theatres/'
theatre_response = requests.get(theatre_URL)
```

```python
type(theatre_response)
```

```python
theatre_response.status_code
```

```python
theatre_response.text
```

```python
theatre_soup = BS(theatre_response.text)
type(theatre_soup)
```

```python
print(theatre_soup.prettify())
```

```python
theatres = theatre_soup.findAll('a', attrs={'class':'hoverable waves-effect'})
```

```python
for house in theatres:
    print(house.text)
```

### scraping list of shows

```python
show_URL = 'https://www.ibdb.com/shows/'
show_response = requests.get(show_URL)
```

```python
type(show_response)
```

```python
show_response.status_code
```

```python
show_response.text
```

```python
show_soup = BS(show_response.text)
type(show_soup)
```

```python
print(show_soup.prettify())
```

```python
shows = show_soup.findAll('i')
```

```python
for production in shows:
    print(production.text)
```
