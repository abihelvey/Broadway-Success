```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
```

```python
gross = pd.read_excel('../data/excel-full-data.xlsx')
gross.head()
```

```python
gross.info()
```

```python
top_10_titles = ['The Lion King', 'Wicked', 'The Phantom of the Opera', 'Hamilton', 'Chicago', 'The Book of Mormon', 'Mamma Mia!', 'Jersey Boys', 'Aladdin', 'Cats']
top_10_grosses = [1740000000, 1430000000, 1300000000, 727840000, 700750000, 687180000, 624400000, 558420000, 492890000, 432000000]
top_10_tickets = [16140000, 12700000, 19600000, 2950000, 9710000, 4320000, 7600000, 5200000, 4630000, 9330000]
top_10_avg_price = [108, 112, 55, 247, 73, 158, 82, 107, 106, 20]
top_10_perf_count = [9600, 7128, 13611, 2205, 9977, 3993, 5772, 4680, 2758, 7401]
top_10_status = ['Open', 'Open', 'Open', 'Open', 'Revived', 'Open', 'Closed', 'Closed', 'Open', 'Closed']
top_10_gross = pd.DataFrame()
top_10_gross['Title'] = top_10_titles
top_10_gross['Total Gross Revenue'] = top_10_grosses
top_10_gross['Total Tickets Sold'] = top_10_tickets
top_10_gross['Average Ticket Price'] = top_10_avg_price
top_10_gross['Performance Count'] = top_10_perf_count
top_10_gross['Open/Closed?'] = top_10_status
top_10_gross
```

```python
top_10_gross.info()
```

```python
#font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}
```

```python
#matplotlib.rc('font', **font)
plt.figure(figsize=(15, 10))
plt.barh(top_10_gross['Title'], top_10_gross['Total Gross Revenue'], color='r')
plt.title('Highest-Grossing Broadway Shows')
plt.xlabel('Total Gross (in billions of dollars)')
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(top_10_gross['Total Gross Revenue'], top_10_gross['Total Tickets Sold'], color='r', s=100)
plt.title('Total Revenue vs. Total Tickets Sold')
plt.ylabel('Total Tickets Sold (in ten-millions)')
plt.xlabel('Total Gross Revenue (in billions of dollars)')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(top_10_gross['Total Gross Revenue'], top_10_gross['Average Ticket Price'], color='r', s=100)
plt.title('Total Revenue vs. Average Ticket Price')
plt.ylabel('Average Ticket Price (in dollars)')
plt.xlabel('Total Gross Revenue (in billions of dollars)')
plt.show()
```
