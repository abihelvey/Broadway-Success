```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
```

```python
gross = pd.read_excel('../data/excel-full-data.xlsx')
gross.head()
```

```python
gross.info()
```

```python
top_10_titles = ['The Lion King', 'Wicked', 'The Phantom of the Opera', 'Hamilton', 'Chicago', 'The Book of Mormon', 'Mamma Mia!', 'Jersey Boys', 'Aladdin', 'Cats']
top_10_grosses = [1740000000, 1430000000, 1300000000, 727840000, 700750000, 687180000, 624400000, 558420000, 492890000, 432000000]
top_10_tickets = [16140000, 12700000, 19600000, 2950000, 9710000, 4320000, 7600000, 5200000, 4630000, 9330000]
top_10_avg_price = [108, 112, 55, 247, 73, 158, 82, 107, 106, 20]
top_10_perf_count = [9600, 7128, 13611, 2205, 9977, 3993, 5772, 4680, 2758, 7401]
top_10_status = ['Open', 'Open', 'Open', 'Open', 'Revived', 'Open', 'Closed', 'Closed', 'Open', 'Closed']
top_10_gross = pd.DataFrame()
top_10_gross['Title'] = top_10_titles
top_10_gross['Total Gross Revenue'] = top_10_grosses
top_10_gross['Total Tickets Sold'] = top_10_tickets
top_10_gross['Average Ticket Price'] = top_10_avg_price
top_10_gross['Performance Count'] = top_10_perf_count
top_10_gross['Open/Closed?'] = top_10_status
top_10_gross
```

```python
top_10_gross.info()
```

```python
plt.figure(figsize=(15, 10))
plt.barh(top_10_gross['Title'], top_10_gross['Total Gross Revenue'], color='r')
plt.title('Highest-Grossing Broadway Shows', fontsize=20)
plt.yticks(fontsize=16)
plt.xlabel('Total Gross (in billions of dollars)', fontsize=16)
plt.xticks(rotation=20, fontsize=14)
#ax.bar_label(ax.containers[0], size=12, padding=3, fmt='%.2f')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(top_10_gross['Total Gross Revenue'], top_10_gross['Total Tickets Sold'], color='r', s=100)
plt.title('Total Revenue vs. Total Tickets Sold', fontsize=20)
plt.ylabel('Total Tickets Sold (in ten-millions)', fontsize=16)
plt.xlabel('Total Gross Revenue (in billions of dollars)', fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(top_10_gross['Total Gross Revenue'], top_10_gross['Average Ticket Price'], color='r', s=100)
plt.title('Total Revenue vs. Average Ticket Price', fontsize=20)
plt.ylabel('Average Ticket Price (in dollars)', fontsize=16)
plt.xlabel('Total Gross Revenue (in billions of dollars)', fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.show()
```

```python
theatres = pd.read_csv('../data/theatre_data.csv')
theatres.head()
```

```python
theatres.info()
```

```python
theatres.sort_values(by=['Capacity'], ascending=False, inplace=True)
theatres.head()
```

```python
theatres.tail()
```

```python
theatres_10big = theatres.iloc[0:10]
theatres_10big.sort_values(by=['Capacity'], ascending=True, inplace=True)
theatres_10big.reset_index(inplace=True)
theatres_10big
```

```python
theatres_10small = theatres.iloc[31:41]
theatres_10small.sort_values(by=['Capacity'], ascending=True, inplace=True)
theatres_10small.reset_index(inplace=True)
theatres_10small
```

```python
plt.figure(figsize=(15,10))
plt.barh(theatres_10big['Theatre'], theatres_10big['Capacity'], color='r')
plt.title('10 Largest Broadway Theatres', fontsize=20)
plt.xlabel('Seat Capacity', fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=14)
plt.axvline(x=500, color='black')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.barh(theatres_10small['Theatre'], theatres_10small['Capacity'], color='r')
plt.title('10 Smallest Broadway Theatres', fontsize=20)
plt.xlabel('Seat Capacity', fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=14)
plt.axvline(x=500, color='black')
plt.show()
```
